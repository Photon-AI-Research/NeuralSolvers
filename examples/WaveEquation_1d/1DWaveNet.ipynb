```python
import sys
sys.path.append('..') # examples
import matplotlib.pyplot as plt
import torch
from NeuralSolvers.models.mlp import MLP
```

```python
x = (-5 - 5) * torch.rand(1000) + 5
t = torch.zeros(1000)
X = torch.stack([x,t],1)
with torch.no_grad():
    label = torch.sin(x)
    label = label.reshape([1000,1])
```

```python
model = MLP(input_size=2, output_size=1, num_hidden=1, hidden_size=100, activation=torch.tanh)
```

```python
def normalize(x):
    lb = torch.tensor([-5,0])
    ub = torch.tensor([5,3])
    return 2.0 * (x - lb) / (ub - lb) - 1.0
    
```

```python
optimizer = torch.optim.Adam(model.parameters(),lr=1e-3)
criterion = torch.nn.MSELoss()
```

```python
for epoch in range(1000):
    optimizer.zero_grad()
    prediction = model(normalize(X))
    loss = criterion(prediction,label)
    loss.backward()
    optimizer.step()
    print("Epoch ", epoch, "Loss ", loss.item())
```

```python
plt.scatter(x,label,label='Ground Truth')
plt.scatter(x,prediction.detach().numpy(),label='Prediction')
plt.legend()
plt.show()
```
